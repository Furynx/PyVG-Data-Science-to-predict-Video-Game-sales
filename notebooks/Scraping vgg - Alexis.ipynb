### Importation des bibliothèques

```python
from bs4 import BeautifulSoup # Importation de BeautifulSoup
import requests # Importation de import requests 
from time import sleep # Importation de sleep
import pandas as pd # Importation de pandas sous l'alias pd
from tqdm import tqdm # Importation de tqdm
import datetime # Importation de datetime
import os # Importation de os
```

### Définition de fonctions

```python
##################################################################
##### Fonction de récupération des données de la page active #####
##################################################################

def retrieve_game_info(link):
    url = 'https://videogamegeek.com/'+link
    user_agent = {'User-agent': 'Mozilla/5.0'}
    response  = requests.get(url, headers = user_agent, timeout=60) # On requête le serveur
    soup = BeautifulSoup(response.text, 'html.parser')
    
    full = []
    for i in range(2,16):
        for item in soup.select('#module_2 tr:nth-child('+str(i)+') td'):
            full.append(item.text.replace('\t','').strip().split('\n'))
            
    if ['Alternate Names'] in full :
        alter= full[full.index(['Alternate Names'])+1]
    else : alter= ['None']
    
    if ['Platform'] in full :
        plat= full[full.index(['Platform'])+1]
    else : plat= ['None']

    if ['Genre'] in full :
        genr= full[full.index(['Genre'])+1]
    else : genr= ['None']
    
    if ['Theme'] in full :
        them= full[full.index(['Theme'])+1]
    else : them= ['None']
        
    if ['Franchise'] in full :
        franch= full[full.index(['Franchise'])+1]
    else : franch= ['None']
        
    if ['Series'] in full :
        seri= full[full.index(['Series'])+1]
    else : seri= ['None']
        
    if ['Mode'] in full :
        mod= full[full.index(['Mode'])+1]
    else : mod= ['None']
 
    if ['Minimum Players'] in full :
        minp= full[full.index(['Minimum Players'])+1]
    else : minp= ['None']

    if ['Maximum Players'] in full :
        maxp= full[full.index(['Maximum Players'])+1]
    else : maxp= ['None']
    
    
    if ['Release Date'] in full :
        rel= full[full.index(['Release Date'])+1]
    else : rel= ['None']

    if ['Developer'] in full :
        dev= full[full.index(['Developer'])+1]
    else : dev= ['None']
    
    if ['Publisher'] in full :
        pub= full[full.index(['Publisher'])+1]
    else : pub= ['None']
    
    if ['Honors'] in full :
        hon= full[full.index(['Honors'])+1]
    else : hon= ['None']
        
    if ['Expansion'] in full :
        exp= full[full.index(['Expansion'])+1]
    else : exp= ['None']
    
    for item in [alter, plat, genr, them, franch, seri, mod, minp, minp, maxp, rel, dev, pub, hon, exp]:
        if len(item)>1: 
            while '' in item:
                del item[item.index('')]
            if 'Show More »' in item : del item[item.index('Show More »')]
            item= ', '.join(item)
        else : item= ''.join(item)

    full = []
    for i in range(1,10):    
        for item in soup.select('#module_17 td:nth-child('+str(i)+') td'):
            full.append(item.text.replace('\t','').strip().split('\n'))           
    nview= full[full.index(['Num Views:'])+1]
    fan= full[full.index(['Fans:'])+1]
    pcom= full[full.index(['Personal\xa0Comments:'])+1]
    uown= full[full.index(['Users\xa0Owning:'])+1]
    uwant= full[full.index(['Users\xa0Wanting:'])+1]
    tplay= full[full.index(['Total\xa0Plays:'])+1]
    mplay= full[full.index(['Plays\xa0This\xa0Month:'])+1]
    
    if len(alter) > 1: alter= ', '.join(alter)
    else : alter= ''.join(alter)
    if len(plat) > 1: plat= ', '.join(plat)
    else : plat= ''.join(plat)
    if len(genr) > 1: genr= ', '.join(genr)
    else : genr= ''.join(genr)
    if len(them) > 1: them= ', '.join(them)
    else : them= ''.join(them)
    if len(franch) > 1: franch= ', '.join(franch)
    else : franch= ''.join(franch)
    if len(seri) > 1: seri= ', '.join(seri)
    else : seri= ''.join(seri)
    if len(mod) > 1: mod= ', '.join(mod)
    else : mod= ''.join(mod)
    if len(minp) > 1: minp= ', '.join(minp)
    else : minp= ''.join(minp)
    if len(maxp) > 1: maxp= ', '.join(maxp)
    else : maxp= ''.join(maxp)
    if len(rel) > 1: rel= ', '.join(rel)
    else : rel= ''.join(rel)
    if len(dev) > 1: dev= ', '.join(dev)
    else : dev= ''.join(dev)
    if len(pub) > 1: pub= ', '.join(pub)
    else : pub= ''.join(pub)
    if len(hon) > 1: hon= ', '.join(hon)
    else : hon= ''.join(hon)
    if len(exp) > 1: exp= ', '.join(exp)
    else : exp= ''.join(exp)
    if len(nview) > 1: nview= ', '.join(nview)
    else : nview= ''.join(nview)
    if len(fan) > 1: fan= ', '.join(fan)
    else : fan= ''.join(fan)
    if len(pcom) > 1: pcom= ', '.join(pcom)
    else : pcom= ''.join(pcom)
    if len(uown) > 1: uown= ', '.join(uown)
    else : uown= ''.join(uown)
    if len(uwant) > 1: uwant= ', '.join(uwant)
    else : uwant= ''.join(uwant)
    if len(tplay) > 1: tplay= ', '.join(tplay)
    else : tplay= ''.join(tplay)
    if len(mplay) > 1: mplay= ', '.join(mplay)
    else : mplay= ''.join(mplay)
    mplay
    
    return alter, plat, genr, them, franch, seri, mod, minp, maxp, rel, dev, pub, hon, exp, nview, fan, pcom, uown, uwant, tplay, mplay  


##################################################################
############## Fonction de récupération du jeu   #################
##################################################################

def retrieve_main_info(soup):
    names=[]
    links=[]
    notes= []
    alters= []
    plats= [] 
    genrs= []
    thems= []
    franchs= []
    seris= []
    mods= []
    minps= []
    maxps= []
    rels= []
    devs= []
    pubs= []
    hons= []
    exps= []
    nviews= []
    fans= []
    pcoms= []
    uowns= []
    uwants= []
    tplays= []
    mplays= []
    
    for item in soup.select('.primary'):
        names.append(item.text)
        links.append(item.attrs['href'])
        lalters, lplats, lgenrs, lthems, lfranchs, lseris, lmods, lminps, lmaxps, lrels, ldevs, lpubs, lhons, lexps, lnviews, lfans, lpcoms, luowns, luwants, ltplays, lmplays = retrieve_game_info(item.attrs['href']) 
        alters.append(lalters)
        plats.append(lplats)
        genrs.append(lgenrs)
        thems.append(lthems)
        franchs.append(lfranchs)
        seris.append(lseris)
        mods.append(lmods)
        minps.append(lminps)
        maxps.append(lmaxps)
        rels.append(lrels)
        devs.append(ldevs)
        pubs.append(lpubs)
        hons.append(lhons)
        exps.append(lexps)
        nviews.append(lnviews)
        fans.append(lfans)
        pcoms.append(lpcoms)
        uowns.append(luowns)
        uwants.append(luwants)
        tplays.append(ltplays)
        mplays.append(lmplays)
        
    for item in soup.select('#row_ .collection_bggrating'):
        notes.append(item.text.strip())
    
    geek_ratings=[notes[i] for i in range(0,len(notes),3)]
    avg_ratings=[notes[i] for i in range(1,len(notes),3)]
    num_votess=[notes[i] for i in range(2,len(notes),3)]
    
    dict_data= {'Name': names,
                'Link': links,
                'Geek_rating': geek_ratings,
                'Avg_rating': avg_ratings,
                'Num_votes': num_votess,
                'Alter_name': alters,
                'Platform': plats,
                'Genre': genrs,
                'Theme': thems,
                'Franchise': franchs,
                'Serie': seris,
                'Mode': mods,
                'Min_players': minps,
                'Max_players': maxps,
                'Release_date': rels,
                'Developer': devs,
                'Publisher': pubs,
                'Honors': hons,
                'Expansion': exps,
                'N_views': nviews,
                'Fans': fans,
                'N_comments': pcoms,
                'Users_owning': uowns,
                'Users_wanting': uwants,
                'Total_plays': tplays,
                'Plays_this_month': mplays                
               }
    
    return dict_data
```

```python
# Initialisation de la page principale de parcours des jeu
vgg_url= 'https://videogamegeek.com/browse/videogame/page/'

# Parcours des pages 1 à 530 et scraping des données si elles n'existent pas
for i in tqdm(range(1,530)):
    filename= 'vgg_'+str(i)+'.csv'
    if os.path.exists(filename) :
        continue
    else :        
        url_page = vgg_url+str(i)
        user_agent = {'User-agent': 'Mozilla/5.0'}
        response  = requests.get(url_page, headers = user_agent, timeout=60) # On requête le serveur
        soup_p = BeautifulSoup(response.text, 'html.parser')
        df=  pd.DataFrame(retrieve_main_info(soup_p))
        df.to_csv(filename, sep=",", encoding='utf-8', index=False) 

# Sauvegarde des données dans un csv     
vgg.to_csv('vgg.csv', sep=",", encoding='utf-8', index=False)
```
