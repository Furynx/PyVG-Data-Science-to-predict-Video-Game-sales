## Scrapping Wikipedia / Mode de jeux : Multiplayer -  Single Player

<img src = './scrapping.png'></img>

```python
#nous allons utiliser l'infobox wikipedia, pour récupérer les informations du mode du jeux : multijouer ou non

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.error import HTTPError

df = pd.read_csv('vgsales.csv',sep = ',')

names = df['Name'].apply(lambda x : x.replace(' ','_'))
names = names.apply(lambda x : x.replace('é','e'))

multiplayer = []
singleplayer  = []
n = []

#parcours des articles wikipédia 
for i in names:
    try:
        
        page = "https://en.wikipedia.org/wiki/"+i
        
        a = urlopen(page)
        soup = BeautifulSoup(a, 'html.parser')
        n.append(i)
        #nous avons itérer le scrapper pour avoir tous les attributs des balises contenant l'information
        #sur le mode de jeux, qui sont les 9 attributs suivants
        cond1= str(soup.findAll(name='a', attrs={'href': '/wiki/Multiplayer_video_game'})) != '[]'
        cond2= str(soup.findAll(name='a', attrs={'href': '/wiki/Two-player_video_game'})) != '[]'
        cond3= str(soup.findAll(name='a', attrs={'href': '/wiki/Multiplayer'}))!= '[]'
        cond4 = str(soup.findAll(name='a', attrs={'href': '/wiki/Single-player'})) != '[]'
        cond5 = str(soup.findAll(name='a', attrs={'href': '/wiki/Single-player_video_game'})) != '[]'
        cond6 = str(soup.findAll(name='a', attrs={'href': '/wiki/Single_player_video_game'})) != '[]'
        cond7 = str(soup.findAll(name='a', attrs={'href': '/wiki/Singleplayer'})) != '[]'
        cond8 = str(soup.findAll(name='a', attrs={'href': '/wiki/Single_Player'})) != '[]'
        cond9 = str(soup.findAll(name='a', attrs={'href': '/wiki/Single_player'})) != '[]'
        
      
        
        multi = True if (cond1 or cond2 or cond3) else False
        single = True if (cond4 or cond5 or cond6 or cond7 or cond8 or cond9) else False
        
        
        
        if (not multi) and (not single):
            multiplayer.append(np.nan)
            singleplayer.append(np.nan)
        else:
            multiplayer.append(multi)
            singleplayer.append(single)
    
    except HTTPError:
        multiplayer.append(np.nan)
        singleplayer.append(np.nan)
    except UnicodeEncodeError:
        print(page)
        multiplayer.append(np.nan)
        singleplayer.append(np.nan)
#ajouts des deux varaibles multiplayer et singlepalyer à notre dataset
df['multiplayer'] = multiplayer
df['singleplayer'] = singleplayer

df.to_csv('vgnew9.csv')
```
