```python
import pandas as pd
#test
```

```python
df1  = pd.read_csv('vgnew_awarded1.csv')
df2 = pd.read_csv('vgsales_1_class.csv')
```

```python
df1 = df1[['Name','multiplayer','singleplayer','awarded']]

df4 = df1.drop_duplicates(subset=['Name'])
```

```python
print(df1.shape)
print(df4.shape)

df3 = df4.merge(right= df2, on= 'Name', how= 'inner')
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df3 = df3.drop(['IP','multiplayer_y','singleplayer_y','videotitle'],axis = 1)
```

```python
df3
```

```python
df3 = df3.dropna()
df3["multiplayer_x"] = df3["multiplayer_x"].astype(int)
df3["singleplayer_x"] = df3["singleplayer_x"].astype(int)
```

```python
df3.info()
```

```python
print(df3.shape)
df3 = df3.drop(['Release_date','Publisher','Developer','Global_Sales','published'],axis = 1)
```

```python
df3.set_index('Name',inplace = True)
```

```python
from sklearn import ensemble
from sklearn.model_selection import train_test_split
df5 = df3.drop('Sales_label_q10',axis = 1)
target = df5.iloc[:,-1]
```

```python
data = df5.drop('Sales_label_q4',axis = 1)

```

```python
data.info()
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,target,test_size = 0.2,random_state = 12)
```

```python
clf= ensemble.RandomForestClassifier(n_jobs = -1,random_state = 321)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test,y_pred,rownames =['reel'],colnames = ['predict'] )
pd.crosstab(y_test, y_pred, rownames=['Classe réelle'], colnames=['Classe prédite'])
```

```python
print('score test',clf.score(X_test,y_test))
print('score train',clf.score(X_train,y_train))

```
