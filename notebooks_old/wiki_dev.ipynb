```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.error import HTTPError
#r1 = re.compile(r">(.*?)<")
df = pd.read_csv('vgnew_series.csv',sep = ',')
names = df['Name'].apply(lambda x : x.replace(' ','_'))
names = names.apply(lambda x : x.replace('é','e'))

developer = []
n = []
nf = ['NO DATA']
for i in names:
    try:
        
        page = "https://en.wikipedia.org/wiki/"+i
        print(page)
        a = urlopen(page)
        soup = BeautifulSoup(a, 'html.parser')
        n.append(i)
        
        
        s_dev = soup.select("tr:nth-child(3) .infobox-data a")
        dev = s_dev[0].text
        developer.append(dev)
        print(dev)

     
        
       
    
    except HTTPError:
        developer.append(np.nan)
       
    except UnicodeEncodeError:
        print(page)
        developer.append(np.nan)
       
    except IndexError:
        developer.append(nf)

df['developer'] = developer


df.to_csv('vgnew_wiki_dev_series.csv')
```
