```python
import wikipediaapi
import pandas as pd
import sys,io
import numpy as np
from  textblob import TextBlob
import spacy
```

```python
df = pd.read_csv('vgnew9.csv',sep = ',')
df2 = pd.DataFrame()
df2['Name'] = df['Name']
names = df['Name'].apply(lambda x : x.replace(' ','_'))
names = names.apply(lambda x : x.replace('é','e'))
```

```python
reception = []
wiki_wiki = wikipediaapi.Wikipedia('en')




for i in names:
    try :
        page = wiki_wiki.page(i)
        x = page.section_by_title('Reception')
        reception.append(x)

    
    
    except IOError :
        reception.append(np.nan)
    except ValueError:
        print("Could not convert data to an integer.")
        reception.append(np.nan)
    except:
        print("Unexpected error:", sys.exc_info()[0])
        reception.append(np.nan)
    

    

df2['reception'] = reception
df2.to_pickle('./reception.pkl')

```

```python
npickled_df = pd.read_pickle("./reception.pkl") 
```

```python
def sentiment_analysis(npickled_df):
    def getSubjectivity(text):
        return TextBlob(text).sentiment.subjectivity
  
 #Create a function to get the polarity
    def getPolarity(text):
       return TextBlob(text).sentiment.polarity
  
 #Create two new columns ‘Subjectivity’ & ‘Polarity’
    npickled_df['TextBlob_Subjectivity'] =    npickled_df['reception'].apply(getSubjectivity)
    npickled_df['TextBlob_Polarity'] = npickled_df['reception'].apply(getPolarity)
    def getAnalysis(score):
        if score < 0:
            return 'Negative'
        elif score == 0:
            return 'Neutral'
        else:
            return 'Positive'
    npickled_df['TextBlob_Analysis'] = npickled_df['TextBlob_Polarity'].apply(getAnalysis)
    return npickled_df
```

```python
npickled_df['reception'] = npickled_df['reception'].astype(str)
```

```python
sentiment_analysis(npickled_df)
```
